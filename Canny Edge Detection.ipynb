```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from scipy import ndimage
```

### Original Image

```python
img = cv2.imread('2.jpg', 0)
img = img.astype('float32')
plt.imshow(img, cmap = 'gray')
plt.show()
```

### Noise Reduction

Edge Detection results are highly sensitive to image noise. To get rid of the noise on the image, I apply Gaussian blur filter to smooth it using ndimage library.

```python
blur = ndimage.gaussian_filter(img, sigma=1)
plt.imshow(blur, cmap = 'gray')
plt.show()
```

### Gradient Calculation

The Gradient calculation step detects the edge intensity and direction by calculating the gradient of the image using edge detection operators. To detect it, the easiest way is to apply sobel filters that highlight this intensity change in both directions. 

#### Sobel X

```python
kx = np.array([[-1,0,+1], [-2,0,+2],  [-1,0,+1]])
x = ndimage.convolve(blur, kx)
plt.imshow(x, cmap = 'gray')
plt.show()
```

#### Sobel Y

```python
ky = np.array([[1,2,1], [0,0,0], [-1,-2,-1]])
y = ndimage.convolve(blur, ky)
plt.imshow(y, cmap = 'gray')
plt.show()
```

#### Magnitude of the gradient

```python
mag = np.hypot(x,y)
mag = mag / mag.max() * 255
plt.imshow(mag, cmap = 'gray')
plt.show()
```

#### The slope theta of the gradient

```python
theta = np.arctan2(y,x)
theta
```

### Non-Maximum Supression

We must perform non-maximum suppression to thin out the edges. The algorithm goes through all the points on the gradient intensity matrix and finds the pixels with the maximum value in the edge directions.

```python
nms = np.zeros((mag.shape), dtype=np.int32)
for i in range(1,mag.shape[0]-1):
    for j in range(1,mag.shape[1]-1):
        try:
            q = 255
            r = 255
            if (0 <= theta[i,j] < 22.5) or (157.5 <= theta[i,j] <= 180):
                q = mag[i, j+1]
                r = mag[i, j-1]
            elif (22.5 <= theta[i,j] < 67.5):
                q = mag[i+1, j-1]
                r = mag[i-1, j+1]
            elif (67.5 <= theta[i,j] < 112.5):
                q = mag[i+1, j]
                r = mag[i-1, j]
            elif (112.5 <= theta[i,j] < 157.5):
                q = mag[i-1, j-1]
                r = mag[i+1, j+1]

            if (mag[i,j] >= q) and (mag[i,j] >= r):
                nms[i,j] = mag[i,j]
            else:
                nms[i,j] = 0

        except IndexError as e:
            pass
```

```python
plt.imshow(nms, cmap = 'gray')
plt.show()
```

### Double Threshold

In this step, we wanto to take the relevant pixels which whill contribute to the final edge. The relevant pixels are strong pixels with high intensity and weak pixels which is not enough to be considered as strong but not small enough to be considered non-relevant. The threshold I used for this assignment is 0.09 for the high threshold ration and 0.05 for the low threshold ratio.

```python
highThreshold = nms.max() * 0.13;
lowThreshold = 0.13 * 0.08;
    
thres = np.zeros((nms.shape), dtype=np.int32)
weak = np.int32(25)
strong = np.int32(255)
    
strong_i, strong_j = np.where(nms >= highThreshold)
weak_i, weak_j = np.where((nms <= highThreshold) & (nms >= lowThreshold))
    
thres[strong_i, strong_j] = strong
thres[weak_i, weak_j] = weak 
```

```python
plt.imshow(thres, cmap = 'gray')
plt.show()
```

### Edge Tracking by Hysteresis

Based on the threshold results, the hysteresis consists of transforming weak pixels into strong ones.

```python
for i in range(1, thres.shape[0]-1):
    for j in range(1, thres.shape[1]-1):
        if (thres[i,j] == weak):
            try:
                if ((thres[i+1, j-1] == strong) or (thres[i+1, j] == strong) or (thres[i+1, j+1] == strong)
                    or (thres[i, j-1] == strong) or (thres[i, j+1] == strong)
                    or (thres[i-1, j-1] == strong) or (thres[i-1, j] == strong) or (thres[i-1, j+1] == strong)):
                    thres[i, j] = strong
                else:
                    thres[i, j] = 0
            except IndexError as e:
                pass
```

```python
plt.imshow(thres, cmap = 'gray')
plt.show()
```

### Canny Edge Detection Result

```python
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(15)

fig.add_subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
fig.add_subplot(1,2,2)
plt.imshow(thres, cmap='gray')
plt.title("Canny Edge Detection")
plt.show(block=True)
```
